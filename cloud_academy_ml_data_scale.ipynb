```python
from google.colab import files
uploaded = files.upload()

# Get the file name (uploaded is a dictionary with filenames as keys)
filename = list(uploaded.keys())[0]

# The file is saved in /content/
filepath = f'/content/{filename}'

print("File path:", filepath)
```

```python
import pandas as pd
boston_df = pd.read_csv(filepath)
boston_df.head()
```

```python
boston_df.describe()
```

```python
X = boston_df.drop('MEDV', axis=1)
y = boston_df['MEDV']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)
X_scaled = pd.DataFrame(X_scaled, columns=list(X))
X_scaled.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
fig,ax = plt.subplots(figsize=(14,8))
ax = sns.boxplot(x='variable', y='value', data=pd.melt(X_scaled))
ax.set_xlabel('')
ax.set_ylabel('MEDV')
plt.show()

```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import RobustScaler

transformer = Pipeline(
    [
      ('scaler', StandardScaler()),
      ('robust', RobustScaler())
    ]

  )

transformer.fit(X)
pd.DataFrame(transformer.transform(X), columns=list(X)).head()
```
