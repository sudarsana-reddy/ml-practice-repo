```python
import pandas as pd
```

```python
from google.colab import files

# Upload file(s)
uploaded = files.upload()

# Get the file name (uploaded is a dictionary with filenames as keys)
filename = list(uploaded.keys())[0]

# The file is saved in /content/
filepath = f'/content/{filename}'

print("File path:", filepath)
```

```python
restuarant_df = pd.read_csv(filepath)
restuarant_df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
ohe_df = pd.DataFrame(ohe.fit_transform(restuarant_df).toarray())
ohe_df.head()
```

```python
pd.DataFrame(ohe.fit_transform(restuarant_df[['smoker']]).toarray()).head()
```

```python
ohe.categories_
```

```python
from sklearn.compose import make_column_transformer
```

```python
ohe = OneHotEncoder()
ct = make_column_transformer(
    (ohe, ['tip', 'city','smoker', 'sex', 'day','time']),
    remainder='passthrough'
)
```

```python
data_new = pd.DataFrame(ct.fit_transform(restuarant_df))
```

```python
data_new.head()
```

```python
#toy dataset
data_toy = pd.DataFrame({
    'constrcutor': ['Ford', 'Tata', 'Toyota']
})
```

```python
ohe_toy = OneHotEncoder()
pd.DataFrame(ohe_toy.fit_transform(data_toy).toarray())
```

```python
new_toy_data = pd.DataFrame({
    'constrcutor': ['Toyota', 'Jeep']
})
```

```python
pd.DataFrame(ohe_toy.transform(new_toy_data).toarray())
```

```python
ohe_toy = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
pd.DataFrame(ohe_toy.fit_transform(data_toy))
```

```python
pd.DataFrame(ohe_toy.transform(new_toy_data))
```
